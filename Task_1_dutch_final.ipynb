# Pre-processing of Data

**Conversion to Lowercase**

```python
def tolower(text):
	return text.lower()

#Demo
tolower("HELLO WORLD")
```

**Removing punctuation**

```python
string.punctuation
```

```python
def removepunctuation(text):
	clean="".join([i for i in text if i not in string.punctuation])
	return clean

removepunctuation("Hi, I'm from CSE!")
```

**Removing Stopwords**

```python
nltk.download('stopwords')
stopwords_list = stopwords.words('dutch')
stopwords_list
```

```python
def removestopwords(text):
	clean = " ".join([word for word in text.split() if word not in stopwords_list])
	return clean

#demo
removestopwords("this weather")
```

**Defining a function to preprocess the text using lambda functions**

```python
df_train['class_label'] = df_train['class_label'].replace(['Yes','No'],[1,0])

def preprocess(df):
	df['tweet_text']= df['tweet_text'].apply(lambda x:tolower(x))
	df['tweet_text']= df['tweet_text'].apply(lambda x:removepunctuation(x))
	df['tweet_text']= df['tweet_text'].apply(lambda x:removestopwords(x))
	return df
```

Applying Preprocessing Functions On Training Data

```python
#Train Data
#Converting the Text to a string and class_label to integer
nltk.download('punkt')
nltk.download('wordnet')
df_train.tweet_text=df_train.tweet_text.astype(str)
df_train.class_label=df_train.class_label.astype(int)
#Applying the function
train = preprocess(df_train)
train_data=train[["tweet_text","class_label"]]
train_data
```

```python
#Test Data
#Converting the Text to a string and class_label to integer
df_test.tweet_text=df_test.tweet_text.astype(str)
#Applying the function
test = preprocess(df_test)
test = test[['tweet_id','tweet_text']]
test
```

# Applying Transformer Models

```python
def predict_result(bert,df):
    predictions, raw_outputs = bert.predict(df['tweet_text'].tolist())

    return predictions
```

**XLM**

```python
def train_xlm_model(model, df):
    xlm = ClassificationModel('xlm', model, num_labels=2, use_cuda=True, args={
        'reprocess_input_data': True,
        'use_cached_eval_features': False,
        'overwrite_output_dir': True,
        'num_train_epochs': 2,
    })
    xlm.train_model(df)
    return xlm
```

```python
class_label_xlm = predict_result(xlm1, test)
class_label_xlm
```

**google-bert/bert-base-multilingual-cased**

```python
from transformers import BertTokenizer, BertModel
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
model = BertModel.from_pretrained("bert-base-multilingual-cased")

from transformers import AutoTokenizer, AutoModel, TFAutoModel

tokenizer = AutoTokenizer.from_pretrained("GroNLP/bert-base-dutch-cased")
model = AutoModel.from_pretrained("GroNLP/bert-base-dutch-cased")  # PyTorch
model = TFAutoModel.from_pretrained("GroNLP/bert-base-dutch-cased")  # Tensorflow

def train_bert_model(model, df):
    bert = ClassificationModel('bert', model, num_labels=2, use_cuda=True, args={
        'reprocess_input_data': True,
        'use_cached_eval_features': False,
        'overwrite_output_dir': True,
        'num_train_epochs': 2,
    })
    bert.train_model(df)
    return bert
```

```python
bert_google1 = train_bert_model('google-bert/bert-base-multilingual-cased', train_data)
```

```python
class_label_google_bert = predict_result(bert_google1, test)
class_label_google_bert
```

**GoNLP**

```python
bert_gonlp1 = train_bert_model("GroNLP/bert-base-dutch-cased", train_data)
```

```python
class_label_gonlp = predict_result(bert_gonlp1, test)
class_label_gonlp
```

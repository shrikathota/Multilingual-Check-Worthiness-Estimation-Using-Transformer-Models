# Pre-processing of Data

**Conversion to Lowercase**

```python
def tolower(text):
	return text.lower()

#Demo
tolower("HELLO WORLD")
```

**Removing punctuation**

```python
string.punctuation
```

```python
def removepunctuation(text):
	clean="".join([i for i in text if i not in string.punctuation])
	return clean

removepunctuation("Hi, I'm from CSE!")
```

**Lemmatisation**

Lemmatisation in NLTK is the algorithmic process of finding the lemma of a word depending on its meaning and context.
It helps in returning the base or dictionary form of a word known as the lemma.

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
e_words = ["wait", "waiting", "waited", "waits"]
for w in e_words:
	print("Lemma for {} is {}".format(w, wordnet_lemmatizer.lemmatize(w)))
```

**Removing Stopwords**

```python
nltk.download('stopwords')
stopwords_list = stopwords.words('english')
stopwords_list
```

```python
def removestopwords(text):
	clean = " ".join([word for word in text.split() if word not in stopwords_list])
	return clean

#demo
removestopwords("this weather")
```

**Defining a function to preprocess the text using lambda functions**

```python
df_train['class_label'] = df_train['class_label'].replace(['Yes','No'],[1,0])

#Lemmetizing
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
def lemmatize_text(text):
    # Tokenize the text into words
    words = word_tokenize(text)
    # Lemmatize each word and join them back into a string
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    return ' '.join(lemmatized_words)

def preprocess(df):
	df['Text']= df['Text'].apply(lambda x:tolower(x))
	df['Text']= df['Text'].apply(lambda x:removepunctuation(x))
	df['Text']= df['Text'].apply(lambda x:removestopwords(x))
	df['Text'] = df['Text'].apply(lemmatize_text)
	return df
```

```python
#Train Data
#Converting the Text to a string and class_label to integer
nltk.download('punkt')
nltk.download('wordnet')
df_train.Text=df_train.Text.astype(str)
df_train.class_label=df_train.class_label.astype(int)
#Applying the function
train = preprocess(df_train)
train_data=train[["Text","class_label"]]
train_data
```

```python
#Test Data
nltk.download('punkt')
nltk.download('wordnet')
#Converting the Text to a string
df_test.Text=df_test.Text.astype(str)
#Applying the function
test = preprocess(df_test)
test
```

# Applying Transformer Models

**BERT**

```python
def predict_result(bert,df):
    predictions, raw_outputs = bert.predict(df['Text'].tolist())
    return predictions
```

```python
def train_bert_model(model,df):
    bert = ClassificationModel('bert', model, num_labels=2, use_cuda=True, args={
    'reprocess_input_data': True,
    'use_cached_eval_features': False,
    'overwrite_output_dir': True,
    'num_train_epochs': 2,
    })
    bert.train_model(df)
    return bert
```

```python
bert1 = train_bert_model('setu4993/LaBSE',train_data)
```

```python
class_label_bert = predict_result(bert1, test)
class_label_bert
```

**DistillBERT**

```python
def train_distilbert_model(model, df):
    distilbert = ClassificationModel('distilbert', model, num_labels=2, use_cuda=True, args={
        'reprocess_input_data': True,
        'use_cached_eval_features': False,
        'overwrite_output_dir': True,
        'num_train_epochs': 2,
    })
    distilbert.train_model(df)
    return distilbert
```

```python
distilbert1 = train_distilbert_model('distilbert-base-uncased', train_data)
```

```python
class_label_distilbert = predict_result(distilbert1, test)
class_label_distilbert
```

**MPNET**

```python
def train_mpnet_model(model, df):
    mpnet = ClassificationModel('mpnet', model, num_labels=2, use_cuda=True, args={
        'reprocess_input_data': True,
        'use_cached_eval_features': False,
        'overwrite_output_dir': True,
        'num_train_epochs': 2,
    })
    mpnet.train_model(df)
    return mpnet
```

```python
mpnet1 = train_mpnet_model('microsoft/mpnet-base', train_data)
```

```python
class_label_mpnet = predict_result(mpnet1, test)
class_label_mpnet
```

# Results

```python
import numpy as np
arr_bert = np.array(class_label_bert)
arr_bert = np.where(arr_bert==0, 'No', 'Yes')

arr_distilbert = np.array(class_label_distilbert)
arr_distilbert = np.where(arr_distilbert==0, 'No', 'Yes')

arr_mpnet = np.array(class_label_mpnet)
arr_mpnet = np.where(arr_mpnet==0, 'No', 'Yes')
```

```python
bert_output = pd.DataFrame(df_test['Sentence_id'])
bert_output.rename(columns = {'Sentence_id':'id'}, inplace = True)
bert_output['class_label'] = arr_bert
bert_output['run_id'] = 'bert'
bert_output
```

```python
distilbert_output = pd.DataFrame(df_test['Sentence_id'])
distilbert_output.rename(columns = {'Sentence_id':'id'}, inplace = True)
distilbert_output['class_label'] = arr_distilbert
distilbert_output['run_id'] = 'distilbert'
distilbert_output
```

```python
mpnet_output = pd.DataFrame(df_test['Sentence_id'])
mpnet_output.rename(columns = {'Sentence_id':'id'}, inplace = True)
mpnet_output['class_label'] = arr_mpnet
mpnet_output['run_id'] = 'mpnet'
mpnet_output
```

```python
output_english = pd.concat([bert_output,distilbert_output,mpnet_output], ignore_index=True, axis=0)
output_english
```
